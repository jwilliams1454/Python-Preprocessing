```python
# import libraries
import numpy as np
import pandas as pd
```

```python
# import the data 
data = pd.read_csv('Nominal.csv')
data
```

```python
# Check the data
data['Pets'].value_counts()
```

```python
# One-Hot Encoding
one_hot_encoded_data = pd.get_dummies(data, columns = ['Color', 'Shape','Pets'])
print(one_hot_encoded_data)
```

```python
#Save dataframe as csv files without index
one_hot_encoded_data.to_csv(r'Encoded.csv', index = False)
```
